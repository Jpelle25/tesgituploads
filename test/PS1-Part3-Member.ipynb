# Super Covid-19 DataFrame 

```python
import pandas as pd

cases_df = pd.read_csv('covid_confirmed_usafacts.csv')
deaths_df = pd.read_csv('covid_deaths_usafacts.csv')
population_df = pd.read_csv('covid_county_population_usafacts.csv')
super_covid_df = pd.merge(cases_df, deaths_df, on = ['countyFIPS', 'County Name', 'State', 'StateFIPS'], suffixes = ['_cases', '_deaths'])
super_covid_df.join(population_df["population"])
display(super_covid_df)
```
