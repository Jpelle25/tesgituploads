# Last week Covid-19 trend

```python
#last_seven = pd.DataFrame(super_covid_df.loc[super_covid_df["State"] == "NV"])
last_seven = pd.DataFrame(super_covid_df.iloc[1771:1789])
last_seven = last_seven[["State", "2023-01-30_x", "2023-01-31_x", "2023-02-01_x", "2023-02-02_x", "2023-02-03_x", "2023-02-04_x", "2023-02-05_x"]]
last_seven = last_seven.reset_index()
last_seven = last_seven.drop(['index'], axis = 1)
data = {
        'State': ['NV'],
        '2023-01-30_cases_sum': [last_seven["2023-01-30_x"].sum()],
        '2023-01-01_cases_sum': [last_seven["2023-01-31_x"].sum()],
        '2023-02-01_cases_sum': [last_seven["2023-02-01_x"].sum()],
        '2023-02-02_cases_sum': [last_seven["2023-02-02_x"].sum()],
        '2023-02-03_cases_sum': [last_seven["2023-02-03_x"].sum()],
        '2023-02-04_cases_sum': [last_seven["2023-02-04_x"].sum()],
        '2023-02-05_cases_sum': [last_seven["2023-02-05_x"].sum()]
       }
data_trend_df = pd.DataFrame(data)
display(last_seven)
display(data_trend_df)
```

#### Trend Evaluation: 

It looks like the trends for NV in the past week have stable for the amount of cases and deaths throughout. Only a few counties experience the slightest variation but nothing that screws the numbers in any direction. Still an extremely high number of cases and deaths but the numbers are consistant throughout the week.
