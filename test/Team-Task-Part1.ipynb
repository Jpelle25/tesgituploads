```python
import pandas as pd
import numpy as np
```

```python
population = pd.read_csv("covid_county_population_usafacts.csv")
confirmed_cases = pd.read_csv("covid_confirmed_usafacts.csv")
deaths = pd.read_csv("covid_deaths_usafacts.csv")
```

```python
population["County Name"] = population["County Name"].str.rstrip()
confirmed_cases["County Name"] = confirmed_cases["County Name"].str.rstrip()
deaths["County Name"] = deaths["County Name"].str.rstrip()
```

```python
def update_county_name(city_name):
    if city_name.endswith("City"):
        return city_name
    elif city_name.startswith("City of"):
        return city_name[8:] + " City"
    elif city_name.startswith("City and County of"):
        return city_name[19:] + " county"
    elif city_name.startswith("City and Borough of"):
        return city_name[20:] + " City and Borough"
    else:
        return city_name

confirmed_cases["County Name"] = confirmed_cases["County Name"].apply(update_county_name)
deaths["County Name"] = deaths["County Name"].apply(update_county_name)
```

```python
def update_county_name_for_population(city_name):
    if city_name.endswith("City"):
        return city_name
    elif city_name.startswith("City of"):
        return city_name[8:] + " City"
    elif city_name.startswith("City and County of"):
        return city_name[19:] + " county"
    elif city_name.startswith("City and Borough of"):
        return city_name[20:] + " City and Borough"
    else:
        return city_name

population["County Name"] = population["County Name"].apply(update_county_name_for_population)
```

```python
confirmed_cases["County Name"] = np.where(confirmed_cases["County Name"] == "Town and County of Nantucket", "Nantucket County", confirmed_cases["County Name"])
confirmed_cases["County Name"] = np.where(confirmed_cases["County Name"] == "Prince of Wales - Hyder Census Area", "Prince of Wales-Hyder Census Area", confirmed_cases["County Name"])
confirmed_cases["County Name"] = np.where(confirmed_cases["County Name"] == "Consolidated Municipality of Carson City", "Carson City", confirmed_cases["County Name"])
confirmed_cases["County Name"] = np.where(confirmed_cases["County Name"] == "Municipality of Skagway", "Skagway Municipality", confirmed_cases["County Name"])

population["County Name"] = np.where(population["County Name"] == "Doña Ana County", "Dona Ana County", population["County Name"])
population["County Name"] = np.where(population["County Name"] == "La Salle Parish", "LaSalle Parish", population["County Name"])  #Broomfield County and City
population["County Name"] = np.where(population["County Name"] == "Broomfield County and City", "Broomfield County", population["County Name"])
population["County Name"] = np.where(population["County Name"] == "Matthews County", "Mathews County", population["County Name"])

deaths["County Name"] = np.where(deaths["County Name"] == "Town and County of Nantucket", "Nantucket County", deaths["County Name"])
deaths["County Name"] = np.where(deaths["County Name"] == "Prince of Wales - Hyder Census Area", "Prince of Wales-Hyder Census Area", deaths["County Name"])
deaths["County Name"] = np.where(deaths["County Name"] == "Consolidated Municipality of Carson City", "Carson City", deaths["County Name"])
deaths["County Name"] = np.where(deaths["County Name"] == "Municipality of Skagway", "Skagway Municipality", deaths["County Name"])
```

```python
population["County Name"] = population["County Name"].str.lower()
confirmed_cases["County Name"] = confirmed_cases["County Name"].str.lower()
deaths["County Name"] = deaths["County Name"].str.lower()
```

```python
cases_and_deaths_df = confirmed_cases.merge(deaths, on=["countyFIPS", "County Name", "State", "StateFIPS"])
```

```python
cases_and_deaths_df = cases_and_deaths_df.merge(population, on=["countyFIPS", "County Name", "State"])
```

```python
cases_and_deaths_df.to_csv("covid19_superdata.csv", index=False)
```
